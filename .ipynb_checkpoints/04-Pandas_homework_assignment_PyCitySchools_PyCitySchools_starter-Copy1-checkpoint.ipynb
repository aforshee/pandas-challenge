```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "04-Pandas_homework_assignment_PyCitySchools_Resources_schools_complete.csv"
student_data_to_load = "04-Pandas_homework_assignment_PyCitySchools_Resources_students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Show data 
school_data_complete.head()
```

```python
#Calculate total number of schools
Total_Schools = len(school_data_complete["school_name"].unique())

#Calculate total number of students
Total_Students = sum(school_data_complete["size"].unique())

#Calculate total budget
Total_Budget = sum(school_data_complete["budget"].unique())

#Calculate math and readding averages
Average_Math = school_data_complete["math_score"].mean()
Average_Reading = school_data_complete["reading_score"].mean()

#Calculate % pass in math
cond = school_data_complete["math_score"] > 69

Number_Pass_Math = school_data_complete["math_score"][cond].count()
Total_Students_Math = (school_data_complete["math_score"]).count()
Percent_Pass_Math = (Number_Pass_Math/ Total_Students_Math) * 100

#Calculate % pass in reading
cond2 = school_data_complete["reading_score"] > 69

Number_Pass_Reading = school_data_complete["reading_score"][cond2].count()
Total_Students_Reading = (school_data_complete["reading_score"]).count()
Percent_Pass_Reading = (Number_Pass_Reading/ Total_Students_Reading)* 100

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Overall = (Average_Math + Average_Reading)/2

#Print results
print(Total_Schools)
print(Total_Students)
print(Total_Budget)
print(Average_Math)
print(Average_Reading)
print(Number_Pass_Math)
print(Total_Students_Math)
print(Percent_Pass_Math)
print(Number_Pass_Reading)
print(Total_Students_Reading)
print(Percent_Pass_Reading)
print(Overall)
```

```python
#Create a dataframe to show results
d = {
    'Total Schools': [Total_Schools], 
    'Total Students': [Total_Students],
    'Total Budget': [Total_Budget],
    'Average Math Score': [Average_Math],
    'Average Reading Score': [Average_Reading],
    '% Passing Math': [Percent_Pass_Math],
    '% Passing Reading': [Percent_Pass_Reading],
    '% Overall Passing Rate': [Overall]
}
Summarydf = pd.DataFrame(data=d)
Summarydf
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Setup conditions for each grade
cond1 = school_data_complete["grade"] == '9th'
cond2 = school_data_complete["grade"] == '10th'
cond3 = school_data_complete["grade"] == '11th'
cond4 = school_data_complete["grade"] == '12th'

# List the columns needed from school_data_complete for the table
cols = ["school_name", "math_score"]

# Use the conditions and cols to create a data series for each grade
Math_9 = school_data_complete[cols][cond1]
Math_10 = school_data_complete[cols][cond2]
Math_11 = school_data_complete[cols][cond3]
Math_12 = school_data_complete[cols][cond4]

# Group each data series by school
grouped_9_Math = Math_9.groupby("school_name")["math_score"].mean()
grouped_10_Math = Math_10.groupby("school_name")["math_score"].mean()
grouped_11_Math = Math_11.groupby("school_name")["math_score"].mean()
grouped_12_Math = Math_12.groupby("school_name")["math_score"].mean()

# Create a dataframe with all of the series
df = pd.DataFrame(columns = ['9th', '10th', '11th', '12th'])
df['9th'] = grouped_9_Math
df['10th'] = grouped_10_Math
df['11th'] = grouped_11_Math
df['12th'] = grouped_12_Math
df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Calculate spending per student by dividing budget by size of school
Per_Student = school_data_complete["budget"]/school_data_complete["size"]

#Create conditional columns on whether or not a student passed math or reading
Pass_Math = np.where(school_data_complete["math_score"] >= 70, 1, 0)
Pass_Read = np.where(school_data_complete["reading_score"] >=70, 1, 0)

#Create % Overall pass rate
Overall = ((Pass_Math + Pass_Read)/2)*100

#Create a df with spending information
With_spending = school_data_complete

#Add calculated columns to with_spending data_frame
With_spending["Average Math Score"] = school_data_complete["math_score"]
With_spending["Average Reading Score"] = school_data_complete["reading_score"]
With_spending["Per Student"] = Per_Student
With_spending["% Passing Math"] = Pass_Math
With_spending["% Passing Reading"] = Pass_Read
With_spending["% Overall Passing Rate"] = Overall
#With_spending["Avg Reading"] = Avg_Reading
#With_spending.head(5)
```

```python
#Name columns that are needed for spending summary table
cols4 = ["Average Math Score", "Average Reading Score", 
         "% Passing Math", "% Passing Reading", "% Overall Passing Rate", "Per Student"]

#Create df with necessary columns
spending_summary = With_spending[cols4]

#Slice data into bins based on spending per student
spending_summary["Spending Ranges (Per Students)"] = pd.cut(spending_summary["Per Student"], 
                                                    spending_bins, labels=group_names, include_lowest=True)

#Group by spending per student bins
grouped_summary = spending_summary.groupby("Spending Ranges (Per Students)")

#Print table with the average of each columns
grouped_summary.mean()
```
